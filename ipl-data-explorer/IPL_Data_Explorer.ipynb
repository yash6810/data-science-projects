```python
#how to check how many match did Virat Kohli played in Ipl

merged_df = pd.merge(df, ddf, left_on='id', right_on='match_id')
kohli_matches = merged_df[(merged_df['batter'] == 'V Kohli') | (merged_df['bowler'] == 'V Kohli')]
kohli_match_ids = kohli_matches['id'].unique()
num_kohli_matches = len(kohli_match_ids)

print(f"Virat Kohli played in {num_kohli_matches} matches.")
```

```python
# Merge the two dataframes on the 'id' and 'match_id' columns
merged_df = pd.merge(df, ddf, left_on='id', right_on='match_id')

# Filter the merged dataframe to include only matches played by Virat Kohli
kohli_matches = merged_df[(merged_df['batter'] == 'V Kohli') | (merged_df['bowler'] == 'V Kohli')]

# Get the season for each unique match Kohli played in
kohli_match_seasons = kohli_matches.drop_duplicates(subset=['id'])['season']

# Count the number of matches per season
kohli_matches_per_season = kohli_match_seasons.value_counts().sort_index()

# Plot the results
plt.figure(figsize=(10, 5))
kohli_matches_per_season.plot(kind='bar')
plt.title('Number of Matches Played by Virat Kohli per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
